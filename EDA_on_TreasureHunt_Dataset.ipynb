**Dropping rows containing Volunteers from the dataset**

```python
form_response[form_response["Final Position"].str.contains("Volunteer")==False]
form_response[form_response["How many team mates did you know before?"].str.contains("volunteers")==False]
```

```python
#Replacing irrevant values with null values


form_response['How many team mates did you know before?'] = form_response['How many team mates did you know before?'].mask(form_response['How many team mates did you know before?'] == 'None').ffill()
form_response['How many team mates did you know before?'] = form_response['How many team mates did you know before?'].mask(form_response['How many team mates did you know before?'] == 'NONE').ffill()
form_response['How many team mates did you know before?'] = form_response['How many team mates did you know before?'].mask(form_response['How many team mates did you know before?'] == 'none').ffill()
form_response['How many team mates did you know before?'] = form_response['How many team mates did you know before?'].mask(form_response['How many team mates did you know before?'] == 'Zero').ffill()
form_response['How many team mates did you know before?'] = form_response['How many team mates did you know before?'].mask(form_response['How many team mates did you know before?'] == '------Null').ffill()
```

```python
# The heat map is useful because it allows us to efficiently grasp
# which variables are strongly correlated with one another.

corrmat = form_response.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=form_response.columns, 
                 xticklabels=form_response.columns, 
                 cmap="Spectral_r")
plt.show()
```

```python
bsc_ea1=form_response['Count of Team Mates from 1 BSc EA'].sum()
bsc_ds1=form_response['Count of Team Mates from 1 BSc DS'].sum()
bsc_ea3=form_response['Count of Team Mates from 3 BSc EA'].sum()
bsc_ds3=form_response['Count of Team Mates from 3 BSc DS'].sum()
bsc_ea5=form_response['Count of Team Mates from 5 BSc EA'].sum()
bsc_ds5=form_response['Count of Team Mates from 5 BSc DS'].sum()
msc = form_response['Count of Team Mates from MSc'].sum()
df = pd.DataFrame({
    'Class': ['1 BSc_EA', '1 BSc DS', '3 BSc EA', '3 BSc DS', '5 BSc EA', '5 BSc DS', 'MSc'],
    'Participant_count': [bsc_ea1, bsc_ds1, bsc_ea3, bsc_ds3, bsc_ea5, bsc_ds5, msc],
})

a = df.plot(title = "Participation from Each Class",x = "Class", y = 'Participant_count', kind = "bar")
```
